# 03B - Learned Baselines

```python
from surprise import Dataset, evaluate, BaselineOnly, Reader

from hyperopt import tpe, fmin, hp, Trials, STATUS_OK

import pickle as pkl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import time

sns.set(style='whitegrid')
% matplotlib inline
```

```python
data_path = os.path.join('..','..','data')
data_file_100 = os.path.join(data_path, 'reviews_sample_100.csv')
data_file_50 = os.path.join(data_path, 'reviews_sample_50.csv')
data_file_25 = os.path.join(data_path, 'reviews_sample_25.csv')
data_file_10 = os.path.join(data_path, 'reviews_sample_10.csv')

df_100 = pd.read_csv(data_file_100).drop('Unnamed: 0', axis = 1)
df_50 = pd.read_csv(data_file_50).drop('Unnamed: 0', axis = 1)
df_25 = pd.read_csv(data_file_25).drop('Unnamed: 0', axis = 1)
df_10 = pd.read_csv(data_file_10).drop('Unnamed: 0', axis = 1)
```

```python
reader = Reader(rating_scale=(1,5))
sdf_100 = Dataset.load_from_df(df_100[['reviewerID', 'asin', 'overall']], reader=reader)
```

```python
def test_baseline(params):
    params['method'] = 'als'
    model = BaselineOnly(bsl_options=params)
    sdf_100.split(3, shuffle=True)
    results = evaluate(model, sdf_100, measures = ['mae'], verbose=True)
    mae = np.array(results['mae']).mean()
    return mae
```

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'trials_baseline_opt'), 'rb') as file_in:
        trials = pkl.load(file_in)
    with open(os.path.join(data_path, 'best_baseline_opt.dict'), 'rb') as file_in:
        best = pkl.load(file_in)
else:
    trials = Trials()
    space = {
        'n_epochs': hp.choice('n_epochs', list(range(5,21))),
        'reg_i': hp.uniform('reg_i', 0, 10),
        'reg_u': hp.uniform('reg_u', 0, 15)
    }
    best = fmin(algo = tpe.suggest,
               fn = test_baseline,
               trials = trials,
               max_evals = 100,
               space = space)
    with open(os.path.join(data_path, 'trials_baseline_opt'), 'wb') as file_out:
        pkl.dump(trials, file_out)
    with open(os.path.join(data_path, 'best_baseline_opt.dict'), 'wb') as file_in:
        pkl.dump(best, file_in)
```

```python
def val_diagnostic(val_name, trials):
    vals = np.array([trial['misc']['vals'][val_name] for trial in trials.trials])
    if val_name == 'n_epochs':
        vals += 5
    elif val_name == 'n_factors':
        vals += 10
    ts = [trial['tid'] for trial in trials.trials]
    results = [trial['result']['loss'] for trial in trials.trials]
    
    fig, axes = plt.subplots(1, 3, figsize = (16,4))
    axes[0].scatter(ts, vals)
    axes[0].set(xlabel='iteration', ylabel=val_name)
    axes[1].hist(np.array(vals).squeeze())
    axes[1].set(xlabel=val_name, ylabel='frequency')
    axes[2].scatter(vals, results)
    axes[2].set(xlabel=val_name, ylabel='loss')
    plt.tight_layout()
```

```python
for val in trials.trials[0]['misc']['vals'].keys():
    val_diagnostic(val, trials)
```

```python
best_opt = best.copy()
best_opt['n_epochs'] += 10
best_opt['method'] = 'als'
best_opt
```

```python
sdf_100.split(5)
model_opt = BaselineOnly(bsl_options=best_opt)
results_opt = evaluate(model_opt, sdf_100, measures = ['mae'])
```

```python
pd.Series(results_opt['mae']).plot(kind = 'box')
```

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'results_baseline_als.array'), 'rb') as file_in:
        results_no_reg_als = pkl.load(file_in)
else:
    best_no_reg_als = best_opt.copy()
    best_no_reg_als['reg_i'] = 0
    best_no_reg_als['reg_u'] = 0
    with open(os.path.join(data_path, 'best_baseline_als.dict'), 'wb') as file_in:
        pkl.dump(best_no_reg_als, file_in)
    model_no_reg_als = BaselineOnly(bsl_options=best_no_reg_als)
    results_no_reg_als = evaluate(model_no_reg_als, sdf_100, measures = ['mae'])
    with open(os.path.join(data_path, 'results_baseline_als.array'), 'wb') as file_out:
        pkl.dump(results_no_reg_als, file_out)
```

```python
pd.Series(results_no_reg_als['mae']).plot(kind = 'box')
```

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'results_baseline_sgd.array'), 'rb') as file_in:
        results_no_reg_sgd = pkl.load(file_in)
else:
    best_no_reg_sgd = best_no_reg_als.copy()
    best_no_reg_sgd['method'] = 'sgd'
    best_no_reg_sgd['n_epochs'] = 30
    with open(os.path.join(data_path, 'best_baseline_sgd.dict'), 'wb') as file_in:
        pkl.dump(best_no_reg_sgd, file_in)
    model_no_reg_sgd = BaselineOnly(bsl_options=best_no_reg_sgd)
    results_no_reg_sgd = evaluate(model_no_reg_sgd, sdf_100, measures = ['mae'])
    with open(os.path.join(data_path, 'results_baseline_sgd.array'), 'wb') as file_out:
        pkl.dump(results_no_reg_sgd, file_out)
```

```python
pd.Series(results_no_reg_sgd['mae']).plot(kind = 'box')
```

# Runtime analysis

```python
df_list = [df_10, df_25, df_50, df_100]

use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'runtime_baseline_als.list'), 'rb') as file_in:
        runtime_als_list = pkl.load(file_in)
else:
    runtime_als_list = list()
    for df in df_list:
        sdf = Dataset.load_from_df(df[['reviewerID', 'asin', 'overall']], reader=reader)
        train = sdf.build_full_trainset()

        start = time.time()
        model_no_reg_als.train(train)
        end = time.time()

        runtime_als_list.append(end-start)

    with open(os.path.join(data_path, 'runtime_baseline_als.list'), 'wb') as file_out:
        pkl.dump(runtime_als_list, file_out)
```

```python
n_reviews_list = [len(df) for df in df_list]
plt.figure(figsize = (10, 5))
plt.plot(n_reviews_list, runtime_als_list)
plt.scatter(n_reviews_list, runtime_als_list, color = 'red')
plt.ylim(0, 5)
plt.xlim(0, 500000)
plt.xlabel('Number of reviews')
plt.ylabel('Training time (seconds)')
plt.title('Training time vs input size')
plt.tight_layout()
```

```python
use_pretrained = True

if use_pretrained:
    with open(os.path.join(data_path, 'runtime_baseline_sgd.list'), 'rb')as file_in:
        runtime_sgd_list = pkl.load(file_in)
else:
    runtime_sgd_list = list()
    for df in df_list:
        sdf = Dataset.load_from_df(df[['reviewerID', 'asin', 'overall']], reader=reader)
        train = sdf.build_full_trainset()

        start = time.time()
        model_no_reg_sgd.train(train)
        end = time.time()

        runtime_sgd_list.append(end-start)

    with open(os.path.join(data_path, 'runtime_baseline_sgd.list'), 'wb')as file_out:
        pkl.dump(runtime_sgd_list, file_out)
```

```python
n_reviews_list = [len(df) for df in df_list]
plt.figure(figsize = (10, 5))
plt.plot(n_reviews_list, runtime_sgd_list)
plt.scatter(n_reviews_list, runtime_sgd_list, color = 'red')
plt.ylim(0, 10)
plt.xlim(0, 500000)
plt.xlabel('Number of reviews')
plt.ylabel('Training time (seconds)')
plt.title('Training time vs input size')
plt.tight_layout()
```
